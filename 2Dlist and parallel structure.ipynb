```python
names = ["Anna" , "Bence", "Csilla", "Dani", "Erika"]
scores = [[78, 84, 91], [87, 83, 89], [75, 81, 88], [88, 82, 87], [79, 86, 90]]
for i in range(0,5):
    print(f"{names[i]}: {scores[i]} ")
means = [np.mean(score) for score in scores]
import matplotlib.pyplot as plt

# Sort the means and names in decreasing order
sorted_indices = np.argsort(means)[::-1]
sorted_means = [means[i] for i in sorted_indices]
sorted_names = [names[i] for i in sorted_indices]

# Use a pretty color palette from seaborn
palette = sns.color_palette("viridis", len(sorted_names))

plt.figure(figsize=(8,5))
bars = plt.bar(sorted_names, sorted_means, color=palette)
plt.title("Average Scores by Name (Decreasing Order)")
plt.xlabel("Name")
plt.ylabel("Average Score")
plt.ylim(0, 100)
plt.show()

```

```python
import hashlib

def name_to_neptun(name):
    """
    Q1: Generate a deterministic 6-character NEPTUN code from a name.
    """
    h = hashlib.sha256(name.encode('utf-8')).hexdigest()
    return h[:6].upper()

# Test Q1
print("Q1 Examples:")
for n in ["Anna", "Bence", "Csilla", "Dani", "Erika", "Gábor"]:
    print(f"{n}: {name_to_neptun(n)}")

def name_and_birth_to_neptun(name, birthdate):
    """
    Q2: Generate a deterministic 6-character NEPTUN code from name and birthdate.
    """
    combined = f"{name}_{birthdate}"
    h = hashlib.sha256(combined.encode('utf-8')).hexdigest()
    return h[:6].upper()

# Test Q2
print("\nQ2 Examples:")
examples_q2 = [
    ("Anna", "2001/05/12"),
    ("Bence", "2002/07/23"),
    ("Csilla", "2000/11/30"),
    ("Dani", "2001/01/15"),
    ("Erika", "2003/03/08"),
    ("Gábor", "1999/09/09"),
    ("Bence", "2004/12/01")  # Same name, different birthdate
]
for n, b in examples_q2:
    print(f"{n} ({b}): {name_and_birth_to_neptun(n, b)}")

def neptun_code(name, birthdate, mothers_name):
    """
    Q3: Generate a deterministic 6-character NEPTUN code from name, birthdate, and mother's name.
    """
    combined = f"{name}_{birthdate}_{mothers_name}"
    h = hashlib.sha256(combined.encode('utf-8')).hexdigest()
    return h[:6].upper()

# Test Q3
print("\nQ3 Examples:")
examples_q3 = [
    ("Anna", "2001/05/12", "Katalin Kovács"),
    ("Anna", "2001/05/12", "Judit Szabó"),  # Same name/birthdate, different mother
    ("Bence", "2002/07/23", "Mária Kiss"),
    ("Bence", "2002/07/23", "Mária Tóth"),  # Same name/birthdate, different mother
    ("Csilla", "2000/11/30", "Ilona Nagy"),
    ("Dani", "2001/01/15", "Ágnes Farkas"),
    ("Erika", "2003/03/08", "Zsuzsa Horváth"),
    ("Gábor", "1999/09/09", "Erzsébet Varga"),
    ("Bence", "2004/12/01", "Mária Kiss"),  # Same mother, different birthdate
]
for n, b, m in examples_q3:
    print(f"{n} ({b}, {m}): {neptun_code(n, b, m)}")

```
