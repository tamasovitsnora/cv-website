def sum_to_n(n):
    """Return the sum of numbers from 1 to n inclusive."""
    return sum(range(1, n + 1))

```python
import yfinance as yf
import matplotlib.pyplot as plt

# List of 5 major oil companies' ticker symbols
oil_tickers = ["XOM", "CVX", "SHEL", "BP", "TOT"]

# Download historical data for all 5 companies from 2024-01-01
oil_data = yf.download(oil_tickers, start="2024-01-01")["Close"]

# Plot the closing prices
plt.figure(figsize=(14, 7))
for ticker in oil_tickers:
    plt.plot(oil_data[ticker], label=ticker)
plt.title("Closing Prices of 5 Major Oil Companies (2024-)")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
normalized = (apple_2025_data['Close'] / apple_2025_data['Close'].iloc[0])
plt.figure(figsize=(12, 6))
plt.plot(normalized)
plt.title("Normalized Apple (AAPL) Closing Price from 2025 (Start = 1)")
plt.xlabel("Date")
plt.ylabel("Normalized Closing Price")
plt.grid(True)
plt.show()
```
